# Removing Outliers

```python
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return lower_bound, upper_bound

def remove_outliers(df, columns):
    df_clean = df.copy()
    for column in columns:
        lower_bound, upper_bound = detect_outliers_iqr(df[column])
        df_clean = df_clean[(df_clean[column] >= lower_bound) & 
                           (df_clean[column] <= upper_bound)]
    return df_clean
```

```python
# Select numerical columns for outlier detection
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Create box plots before outlier removal
plt.figure(figsize=(15, 5))
sns.boxplot(data=df[numerical_columns])
plt.title('Distribution Before Outlier Removal')
plt.xticks(rotation=45)
plt.show()
```

```python
# Remove outliers
df_clean = remove_outliers(df, numerical_columns)

# Create box plots after outlier removal
plt.figure(figsize=(15, 5))
sns.boxplot(data=df_clean[numerical_columns])
plt.title('Distribution After Outlier Removal')
plt.xticks(rotation=45)
plt.show()

# Print the number of rows removed
print(f"Original dataset size: {len(df)}")
print(f"Dataset size after outlier removal: {len(df_clean)}")
print(f"Number of rows removed: {len(df) - len(df_clean)}")
```

```python
dine_in_stats = df.groupby('dine_in')['minimum_delivery_time'].describe()
plt.figure(figsize=(14, 8))
sns.boxplot(x='dine_in', y='minimum_delivery_time', data=df, palette='Set3')
plt.title('Minimum Delivery Time by Dine-In Option', fontsize=16)
plt.xlabel('Dine-In Option', fontsize=14)
plt.ylabel('Minimum Delivery Time (minutes)', fontsize=14)
plt.xticks([0, 1], ['No Dine-In', 'Dine-In'], fontsize=12)
for i, dine_in in enumerate(dine_in_stats.index):
    stats = dine_in_stats.loc[dine_in]
    textstr = '\n'.join((
        f"Count: {int(stats['count']):,}",
        f"Mean: {stats['mean']:.2f}",
        f"Std Dev: {stats['std']:.2f}",
        f"Median: {stats['50%']:.2f}",
        f"Min: {stats['min']:.2f}",
        f"Max: {stats['max']:.2f}"
    ))
    plt.text(i, stats['max'], textstr, ha='center', va='top', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))

plt.tight_layout()
plt.show()
```

```python
rating_stats = df.groupby('dine_in')['rating'].describe()

plt.figure(figsize=(14, 8))
sns.boxplot(x='dine_in', y='rating', data=df, palette='Set3')
plt.title('Ratings by Dine-In Option', fontsize=16)
plt.xlabel('Dine-In Option', fontsize=14)
plt.ylabel('Rating', fontsize=14)
plt.xticks([0, 1], ['No Dine-In', 'Dine-In'], fontsize=12)

for i, dine_in in enumerate(rating_stats.index):
    stats = rating_stats.loc[dine_in]
    textstr = '\n'.join((
        f"Count: {int(stats['count']):,}",
        f"Mean: {stats['mean']:.2f}",
        f"Std Dev: {stats['std']:.2f}",
        f"Median: {stats['50%']:.2f}",
        f"Min: {stats['min']:.2f}",
        f"Max: {stats['max']:.2f}"
    ))
    plt.text(i, stats['max'], textstr, ha='center', va='top', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))

plt.tight_layout()
plt.show()
```
